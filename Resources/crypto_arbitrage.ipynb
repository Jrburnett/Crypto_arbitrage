#### 1. For each of the three dates, measure the arbitrage spread between the two exchanges by subtracting the lower-priced exchange from the higher-priced one. *NOTE*: For illustration, only one of the three dates is shown in the starter code below.

```python
# For the date early in the dataset, measure the arbitrage spread between the two exchanges
# by subtracting the lower-priced exchange from the higher-priced one
arbitrage_spread_early = coinbase['Close'].loc['2018-01-01'] - bitstamp['Close'].loc['2018-01-01']
arbitrage_spread_middle = coinbase['Close'].loc['2018-02-02'] - bitstamp['Close'].loc['2018-02-02']
arbitrage_spread_late = coinbase['Close'].loc['2018-03-03'] - bitstamp['Close'].loc['2018-03-03']
# Use a conditional statement to generate the summary statistics for each arbitrage_spread DataFrame
# YOUR CODE HERE
# early arbitrage spread summary statistics
profitable_spread_early = arbitrage_spread_early[arbitrage_spread_early > 0]
profitable_spread_early.describe()
```

```python
# middle arbitrage summary statistics
profitable_spread_middle = arbitrage_spread_middle[arbitrage_spread_middle > 0]
profitable_spread_middle.describe()
```

```python
# late arbitrage summary statistics
profitable_spread_late = arbitrage_spread_late[arbitrage_spread_late > 0]
profitable_spread_late.describe()
```

#### 2. For each of the three dates, calculate the spread returns. To do so, divide the instances that have a positive arbitrage spread (that is, a spread greater than zero) by the price of Bitcoin from the exchange you’re buying on (that is, the lower-priced exchange). Review the resulting DataFrame.

```python
# For the date early in the dataset, calculate the spread returns by dividing the instances when the arbitrage spread is positive (> 0) 
# by the price of Bitcoin from the exchange you are buying on (the lower-priced exchange).
spread_return_early = profitable_spread_early / bitstamp['Close'].loc['2018-01-01']

# Review the spread return DataFrame
# YOUR CODE HERE
spread_return_early.describe()
```

```python
# middle dataset caculating the spread returns by dividing the instance when the abritage spread is positive
spread_return_middle = profitable_spread_middle / bitstamp['Close'].loc['2018-02-02']

spread_return_middle.describe()
```

```python
# late dataset caculating the spread returns by dividing the instances when the abritage spread is positive
spread_return_late = profitable_spread_late / bitstamp['Close'].loc['2018-03-03']

spread_return_late.describe()
```
